```julia
#Problem 1: If we list all the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6 and 9. 
#The sum of these multiples is 23.
#Find the sum of all the multiples of 3 or 5 below 1000.
x = 0
for i in collect(1:999)
    if i%3 == 0 || i%5 == 0
        x += i
    end
end
print(x)
```

```julia
# Problem 2. By considering the terms in the Fibonacci sequence whose values 
#do not exceed four million, find the sum of the even-valued terms.
function fibonacci(n)
    if n < 2 
        return 1
    end
    fibonacci(n-1) + fibonacci(n-2)
end


a=0
x = fibonacci.(1:40)
for i in x
    if i%2 == 0 && i < 4000000
        a+=i
    end
end
println(a)
```

```julia
#Problem 4: A palindromic number reads the same both ways. 
#The largest palindrome made from the product of two 2-digit numbers is 9009 = 91  99.
#Find the largest palindrome made from the product of two 3-digit numbers.


for i in reverse(collect(900:999))
    for j in reverse(collect(900:999))
        x = i*j
        y = string(x)
        if y[1] == y[6] && y[2] == y[5] && y[3]==y[4]
            return i, j, x
        end
    end
end
```

```julia
#problem 12: What is the value of the first triangle number to have over five hundred divisors?
function get_factors(x)
    result = []
    for j in 1:((x/2)+1)
        if x%j == 0
            append!(result, Int(j))
        end
    end
    append!(result, x)
    return(result)
end

b = 0
for i in 1:100000000
    a = b + i
    b = a
    if (a % 2 == 0) & (a % 3 == 0) & (a % 5 ==0) & (a % 7 == 0)
        len = length(get_factors(a))
    else
        continue
    end
    if len >= 500
        println(a)
        break
    end
end
```
